```python
new_df=realestate[[
                  'RegionName',
                'State',
                  '2000',
                  '2001',
                  '2002',
                  '2003',
                  '2004',
                  '2005',
                  '2006',
                  '2007',
                  '2008',
                  '2009',
                  '2010',
                  '2011',
                  '2012',
                  '2013',
                  '2014',
                  '2015',
                  '2016',
                  '2017',
                  '2018']].copy()
new_df.head()
```

```python
new_df = pd.melt(new_df, 
        id_vars=['RegionName','State'], 
        value_vars=['2000', '2001',"2002","2003","2004","2005","2006",
                    "2007","2008","2009","2010","2011","2012","2013",
                    "2014","2015","2016","2017","2018"])
```

```python
new_df = new_df.rename(columns={
    'RegionName':"County",
    "variable":"Year",
    "value":"Price"
})
new_df
```

```python
new_df = new_df.loc[new_df["State"].isin(["DC","MD","VA","WA"])]
filtered_df = new_df.loc[new_df["County"].isin([
'Montgomery',
'Prince Georges',
'District of Columbia',
'Arlington',
'Frederick',
'Charles',
'Clarke',
'Culpeper',
'Fairfax County',
'Fairfax City',
'Falls Church City',
'Fauquier',
'Calvert',
'Fredericksburg City',
'Jefferson',
'Loudoun',
'Manassas Park City',
'Prince William',
'Spotsylvania',
'Stafford',
'Warren',
"Alexandria"])]
filtered_df
```

```python
filtered_df.count()
```

```python
train_years = ["2000","2001","2002", "2003","2004","2005","2006","2007","2008", "2009", "2010"]

test_years = ["2011","2012","2013","2014","2015","2016","2017","2018"]

print(train_years)
print(test_years)
```

```python
trainingdata_df = filtered_df.loc[filtered_df['Year'].isin(train_years)]

trainingdata_df
```

```python
trainingdata_df.count()
```

```python
testingdata_df = filtered_df.loc[filtered_df['Year'].isin(test_years)]

testingdata_df
```

```python
testingdata_df.count()
```

```python
trainingdata_all_df = new_df.loc[new_df['Year'].isin(train_years)]

trainingdata_all_df.count()
```

```python
testingdata_all_df = new_df.loc[new_df['Year'].isin(test_years)]

testingdata_all_df.count()
```
